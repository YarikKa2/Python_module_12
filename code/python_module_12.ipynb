```python
import os
path = os.path.dirname(os.getcwd())
path = path.replace('\\', '/') + '/'
print(path)
```

```python
import pandas as pd
melb_data = pd.read_csv(path + 'CSVs/melb_data_fe.csv', sep=',')
display(melb_data.head())
```

```python
# преобразование столбцов к нужным типам
melb_data['Date'] = pd.to_datetime(melb_data['Date'])

type_obj_mask = melb_data.select_dtypes('object') # выбираем объекты класса object
col_names = type_obj_mask.columns # создаем список с названиями колонок

melb_data[col_names] = type_obj_mask.apply(lambda x: x.astype('category')) # проходимся по колонкам с типом object и заменяем object на category 
melb_data.info()
```

```python
# сортировка значений
melb_data.sort_values(by='Price').head(10) # сортируем значения по возрастанию
melb_data.sort_values(by='Date', ascending=False).head(10) # сортируем значение по убывнию
melb_data.sort_values(by=['SellerG', 'Price'], ascending=True).head(10) # сортируем значения по возрастанию по двум признакам
```

```python
# Предположим, компания McGrath поручила нам восстановить хронологию продаж таунхаусов, 
# у которых площадь участка существенно больше площади здания, чтобы понять, как часто компания справляется с таким сложным видом объектов. 
# Объекты, проданные в один и тот же день, мы бы хотели сортировать по значению коэффициента соотношения площадей. 


# создаем 3 условия
area_ration_mask = melb_data['AreaRatio'] < -0.8    # значение AreaRatio очень низкое 
type_mask = melb_data['Type'] == 'townhouse'        # тип таунхауз
company_mask = melb_data['SellerG'] == 'McGrath'    # продавец McGrath

# при выборе данных вышелобозначенные условия перечисляем через & (НЕ ЗАПЯТАЯ) 
melb_data[area_ration_mask & company_mask & type_mask].sort_values(by=['Date', 'AreaRatio'], # сортируем
                                                                   ascending=True, ignore_index=True # в возрастающем порядке 
                                                                   ).loc[:, ['Date', 'AreaRatio']]  # выводим только столбцы 'Date' и 'AreaRatio'
```

```python
# группировка groupby

display(melb_data.groupby(by='Type', as_index=False,  observed=False).mean(numeric_only=True))

# Давайте сравним средние цены на объекты в зависимости от их типа:
melb_data.groupby('Type',  observed=False)['Price'].mean()

# Теперь давайте выясним, какие регионы (Regionname) наиболее удалены от центра Мельбурна.
melb_data.groupby('Regionname',  observed=False)['Distance'].min().sort_values(ascending=False)

# Чтобы рассчитать несколько агрегирующих методов, можно воспользоваться методом agg(), который принимает список строк с названиями агрегаций.
melb_data.groupby('MonthSale')['Price'].agg('describe').sort_values(by='count', ascending=True) # agg('describe') выведет ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
melb_data.groupby('MonthSale')['Price'].agg(['count', 'mean', 'max']).sort_values(by='count', ascending=True)

# Мы можем вычислить число уникальных риелторских компаний в зависимости от региона, чтобы понять, в каких регионах конкуренция на рынке недвижимости меньше.
melb_data.groupby('Regionname')['SellerG'].agg(['nunique', set]).sort_values(by='nunique')
```

```python
# Построение сводных таблиц groupby

# Можно построить таблицу, в которой мы будем учитывать не только число комнат, но и тип здания (Type). 
# Для этого в параметрах метода groupby() укажем список из нескольких интересующих нас столбцов.

melb_data.groupby(['Rooms', 'Type'])['Price'].mean()

# Для того, чтобы финальный результат был представлен в виде сводной таблицы используют метод unstack()

melb_data.groupby(['Rooms', 'Type'])['Price'].mean().unstack()


# Построение сводных таблиц pivot_table 

melb_data.pivot_table(
    values='Price', # имя столбца из которого берем значения
    index='Rooms', # имя столбца который станет строками
    columns='Type', # имя столбца который станет столбцами
    fill_value=0,  # на что заменить NaN
    aggfunc='mean' # функция, которую применяем (по умолчанию mean)
                   # допустимо применять несколько функций             
).round()


melb_data.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)


#  Многомерные сводные таблицы
melb_data.pivot_table(
    values='Price',
    index=['Method','Type'], # несколько индексов
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)


pivot = melb_data.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'], # несколько функций
    fill_value=0
)

display(pivot)

pivot.columns

# Из таблицы pivot мы можем получить средние значения площадей участков для типа здания unit, 
# просто последовательно обратившись по имени столбцов:
display(pivot['mean']['unit'])

# Аналогично производится и фильтрация данных. Например, если нам нужны регионы, 
# в которых средняя площадь здания для домов типа house меньше их медианной площади, то мы можем найти их следующим образом:

pivot_mask = pivot['mean']['house'] < pivot['median']['house']
pivot_mod = pivot[pivot_mask]
display(pivot_mod)
print(list(pivot_mod.index)) # вывод только Regionname
```

```python
# В этой части модуля мы будем работать с популярным датасетом MovieLens
movies_data = pd.read_csv(path + 'CSVs/movies_data/movies.csv', sep=',')
ratings_1 = pd.read_csv(path + 'CSVs/movies_data/ratings1.csv', sep=',')
ratings_2 = pd.read_csv(path + 'CSVs/movies_data/ratings2.csv', sep=',')
dates_data = pd.read_csv(path + 'CSVs/movies_data/dates.csv', sep=',')
dates_data['date'] = pd.to_datetime(dates_data['date'])

# конкатенируем (склеиваем) таблицы
ratings = pd.concat(
    [ratings_1, ratings_2],
    ignore_index=True) # ставим на  True, чтобы не было дубликатов айдишников


# удаляем дубликаты
ratings = ratings.drop_duplicates(ignore_index=True)


# склеиваем таблицы по столбам
ratings_date = pd.concat(
    [ratings, dates_data],
    axis=1
    
)
display(ratings_date)
```

```python
# Метод объединения join

joined_false = ratings.join(
    movies_data,
    rsuffix='_right',
    how='left'
)


joined  = ratings_date.join(
    movies_data.set_index('movieId'),
    on='movieId',
    how='left'
)

display(joined_false)
display(joined)
```

```python
# Метод объединения merge

merged = ratings_date.merge(
    movies_data,
    on='movieId',
    how='left'
)

display(merged)

# метод merged автоматически удаляет дубликаты и игнорирует индексы
ratings_merged = ratings_1.merge(
    ratings_2,
    how='outer')

display(ratings_merged)
```
