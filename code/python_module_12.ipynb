```python
import os
path = os.path.dirname(os.getcwd())
path = path.replace('\\', '/') + '/'
print(path)
```

```python
import pandas as pd
melb_data = pd.read_csv(path + 'CSVs/melb_data_fe.csv', sep=',')
display(melb_data.head())
```

```python
# преобразование столбцов к нужным типам
melb_data['Date'] = pd.to_datetime(melb_data['Date'])

type_obj_mask = melb_data.select_dtypes('object') # выбираем объекты класса object
col_names = type_obj_mask.columns # создаем список с названиями колонок

melb_data[col_names] = type_obj_mask.apply(lambda x: x.astype('category')) # проходимся по колонкам с типом object и заменяем object на category 
melb_data.info()
```

```python
# сортировка значений
melb_data.sort_values(by='Price').head(10) # сортируем значения по возрастанию
melb_data.sort_values(by='Date', ascending=False).head(10) # сортируем значение по убывнию
melb_data.sort_values(by=['SellerG', 'Price'], ascending=True).head(10) # сортируем значения по возрастанию по двум признакам
```

```python
# Предположим, компания McGrath поручила нам восстановить хронологию продаж таунхаусов, 
# у которых площадь участка существенно больше площади здания, чтобы понять, как часто компания справляется с таким сложным видом объектов. 
# Объекты, проданные в один и тот же день, мы бы хотели сортировать по значению коэффициента соотношения площадей. 


# создаем 3 условия
area_ration_mask = melb_data['AreaRatio'] < -0.8    # значение AreaRatio очень низкое 
type_mask = melb_data['Type'] == 'townhouse'        # тип таунхауз
company_mask = melb_data['SellerG'] == 'McGrath'    # продавец McGrath

# при выборе данных вышелобозначенные условия перечисляем через & (НЕ ЗАПЯТАЯ) 
melb_data[area_ration_mask & company_mask & type_mask].sort_values(by=['Date', 'AreaRatio'], # сортируем
                                                                   ascending=True, ignore_index=True # в возрастающем порядке 
                                                                   ).loc[:, ['Date', 'AreaRatio']]  # выводим только столбцы 'Date' и 'AreaRatio'
```

```python
# группировка groupby

display(melb_data.groupby(by='Type', as_index=False,  observed=False).mean(numeric_only=True))

# Давайте сравним средние цены на объекты в зависимости от их типа:
melb_data.groupby('Type',  observed=False)['Price'].mean()

# Теперь давайте выясним, какие регионы (Regionname) наиболее удалены от центра Мельбурна.
melb_data.groupby('Regionname',  observed=False)['Distance'].min().sort_values(ascending=False)

# Чтобы рассчитать несколько агрегирующих методов, можно воспользоваться методом agg(), который принимает список строк с названиями агрегаций.
melb_data.groupby('MonthSale')['Price'].agg('describe').sort_values(by='count', ascending=True) # agg('describe') выведет ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
melb_data.groupby('MonthSale')['Price'].agg(['count', 'mean', 'max']).sort_values(by='count', ascending=True)

# Мы можем вычислить число уникальных риелторских компаний в зависимости от региона, чтобы понять, в каких регионах конкуренция на рынке недвижимости меньше.
melb_data.groupby('Regionname')['SellerG'].agg(['nunique', set]).sort_values(by='nunique')
```
